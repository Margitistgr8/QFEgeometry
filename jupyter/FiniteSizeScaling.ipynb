```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("/Users/jinyunlin/QFE-Research/QFEgeometry/data/IsingCT/IsingCT.dat")
dataN8 = data[data[:, 0]==8]
dataN16 = data[data[:, 0]==16]
dataN24 = data[data[:, 0]==24]
dataN32 = data[data[:, 0]==32]
dataN40 = data[data[:, 0]==40]

tcoup8 = dataN8[:,3]
tcoup16 = dataN16[:,3]
tcoup24 = dataN24[:,3]
tcoup32 = dataN32[:,3]
tcoup40 = dataN40[:,3]
sind8 = np.argsort(tcoup8)
sind16 = np.argsort(tcoup16)
sind24 = np.argsort(tcoup24)
sind32= np.argsort(tcoup32)
sind40= np.argsort(tcoup40)

tcoup8 = dataN8[:,3][sind8]
tcoup16 = dataN16[:,3][sind16]
tcoup24 = dataN24[:,3][sind24]
tcoup32 = dataN32[:,3][sind32]
tcoup40 = dataN40[:,3][sind40]

m8 = dataN8[:,6][sind8]
m16 = dataN16[:,6][sind16]
m24 = dataN24[:,6][sind24]
m32 = dataN32[:,6][sind32]
m40 = dataN40[:,6][sind40]

m28 = dataN8[:,8][sind8]
m216 = dataN16[:,8][sind16]
m224 = dataN24[:,8][sind24]
m232 = dataN32[:,8][sind32]
m240 = dataN40[:,8][sind40]

U8 = dataN8[:,-1][sind8]
U16 = dataN16[:,-1][sind16]
U24 = dataN24[:,-1][sind24]
U32 = dataN32[:,-1][sind32]
U40 = dataN40[:,-1][sind40]
```

```python
# plt.scatter(tcoup8, m8)
# plt.scatter(tcoup16, m16)
plt.plot(tcoup24, m24)
plt.plot(tcoup32, m32)
plt.plot(tcoup40, m40)
```

```python
#plt.scatter(tcoup8, m28-m8*m8)
plt.scatter(tcoup16, m216-m16*m16)
plt.scatter(tcoup24, m224-m24*m24)
plt.scatter(tcoup32, m232-m32*m32)
plt.scatter(tcoup40, m240-m40*m40)
```

```python
# plt.plot(tcoup8, U8)
# plt.plot(tcoup16, U16)
plt.plot(tcoup24, U24)
plt.plot(tcoup32, U32)
plt.plot(tcoup40, U40)
plt.xlim(2.5, 3.1)
```
